```python
import argparse
from datetime import datetime
from deepreg.train import train

name = "paired_mrus_brain3"

log_root = f""
log_dir = "logs_train/" + datetime.now().strftime("%Y%m%d-%H%M%S")
config_path = [f"{name}.yaml"]
```

```python
history = train(
    gpu="2", # gpu="0,1,2,3" for 4 GPUs
    config_path=config_path,
    gpu_allow_growth=False,
    ckpt_path="",
    log_dir=log_dir,
)
```

```python
min(history.history['val_loss/label_L2_Distance'])
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss/label_L2_Distance'], label = 'Training Loss')
plt.plot(history.history['val_loss/label_L2_Distance'], label = 'Validation Loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('L2 Loss')
```
